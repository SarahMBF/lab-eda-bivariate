```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
#Part 1: Analyzing Best-Seller Trends Across Product Categories
```

```python
# Loading the housing price dataset (assuming the file name is "housing_price.csv")
df = pd.read_csv(r"C:\Users\sarah\OneDrive\Documentos\GitHub\lab-eda-bivariate\amz_uk_price_prediction_dataset.csv")
```

```python
df
```

```python
# Generating a crosstab for 'MSZoning' and 'SaleCondition'
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])

crosstab_result
```

```python
crosstab_first = crosstab_result.copy()
```

```python
crosstab_first['best_seller_ratio'] = (crosstab_result[True] / (crosstab_result[False] + crosstab_result[True]))
```

```python
crosstab_first.sort_values('best_seller_ratio', ascending=False, inplace=True)
crosstab_first
```

```python
# Statistical Analysis
from scipy.stats import chi2_contingency

chi2_statistic, chi2_p_value, _, _ = chi2_contingency(crosstab_result)

chi2_statistic, chi2_p_value
```

```python
# To make it more visually appealing I will only show the top 10 values.
crosstab_result.sort_values(True, ascending = False).head(10).plot(kind="bar", stacked=True)
plt.show()
```

```python
#Part 2: Exploring Product Prices and Ratings Across Categories and Brands
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
            
            # Define bounds for the outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
            
            
df_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
    
```

```python
df_outliers
```

```python
# Violin plots

sns.violinplot(data=df_outliers, x='category', y='price')
plt.show()
```

```python
top20_cat = df_outliers['category'].value_counts().head(20).index
```

```python
df_top20 = df_outliers[df_outliers['category'].isin(top20_cat)]
```

```python
# Create violin plot
sns.violinplot(data=df_top20, x='category', y='price')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# median price per category
median_prices = df_outliers.groupby('category')['price'].median().sort_values(ascending=False)
median_prices
```

```python
median_prices = df_outliers.groupby('category')['price'].mean().sort_values(ascending=False)
median_prices
```

```python
highest_median_value = median_prices.max()
highest_median_value
```

```python
# Bar Charts:
top10_cat = df_outliers['category'].value_counts().head(10).index
df_top = df_outliers[df_outliers['category'].isin(top10_cat)]
df_top
```

```python
df_top10 = pd.DataFrame(df_top.groupby('category')['price'].mean()).reset_index()
```

```python
df_top10
```

```python
sns.barplot(data=df_top10, x='category', y='price', palette="coolwarm")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
df_top = df_top[df_top['stars'] > 0]
```

```python
sns.boxplot(data=df_top, x='category', y='stars')
plt.xticks(rotation=45, ha='right')
plt.show()
```
